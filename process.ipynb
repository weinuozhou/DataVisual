1. 将基础数据保存到`data.json`
    * 不同门店的销售额和利润额
    * 不同商品类别（包括子类别）产生的销售额和利润额
    * 不同年份的销售额与利润额
    * 不同地区（东北、中南...）的销售额与利润额
    * 不同支付方式的数量
    * 不同客户类型的数量
2. 将地图相关数据保存到`area.json`
    * 中国每年的各个省份的销售额与利润额（注意有缺失值，需要进一步处理）

```python
import pandas as pd
```

```python
df = pd.read_excel("./data/商品订单表.xlsx")
df.head()
```

```python
totalSale = df["销售额"].sum()
totalProfit = df["利润额"].sum()
rate = totalProfit / totalSale
df2 = pd.pivot_table(df, values=["销售额", "利润额"], index="门店名称", aggfunc="sum")
StoreName = df2.index.values
saleByStore = df2["销售额"].values
profitByStore = df2["利润额"].values
df3 = pd.pivot_table(df, values=["销售额", "利润额"], index="商品类别", aggfunc="sum")
category = df3.index
saleByCategory = df3["销售额"].values
profitByCategory = df3["利润额"].values
df4 = pd.pivot_table(df, values=["销售额", "利润额"], index=["商品类别", "子类别"], aggfunc="sum")
subcategory = [i[1] for i in df4.index]
saleBySubcategory = df4["销售额"].values
profitBySubcategory = df4["利润额"].values
payWay = df["支付方式"].value_counts().index
payCount = [i for i in df["支付方式"].value_counts()]
customer = df["客户类型"].value_counts().index
customerCount = [i for i in df["客户类型"].value_counts()]
df5 = pd.pivot_table(df, values=["销售额", "利润额"], index="年份", aggfunc="sum")
years = df5.index
saleByYear = df5["销售额"].values
profitByYear = df5["利润额"].values
df6 = pd.pivot_table(df, values=["销售额", "利润额"], index="地区", aggfunc="sum")
regions = df6.index
saleByRegion = df6["销售额"].values
profitByRegion = df6["利润额"].values
df7 = pd.pivot_table(df, values=["销售额", "利润额"], index="销售经理", aggfunc="sum")
df7.sort_values(["销售额", "利润额"], ascending=[False, False], inplace=True)
like_name = df7.index[0]
saleByLike = df7["销售额"][0]
profitByLike = df7["利润额"][0]
```

```python
data = {
    "info": {
        "totalSale": totalSale,
        "totalProfit": totalProfit,
        "rate": rate
    },
    "store": {
        "storeName": list(StoreName),
        "saleByStore": list(saleByStore),
        "profitByStore": list(profitByStore)
    },
    "category":{
        "name": list(category),
        "saleByCategory": list(saleByCategory),
        "profitByCategory": list(profitByCategory),
        "subcategory": {
            "name": subcategory,
            "saleBySubcategory": list(saleBySubcategory),
            "profitBySubcategory": list(profitBySubcategory)
        }
    },
    "pay": {
        "payWay": list(payWay),
        "payCount": payCount
    },
    "customer": {
        "type": list(customer),
        "customerCount": customerCount
    },
    "year":{
        "years": list(years),
        "saleByYear": list(saleByYear),
        "profitByYear": list(profitByYear)
    },
    "region":{
        "regionName": list(regions),
        "saleByRegion": list(saleByRegion),
        "profitByRegion": list(profitByRegion)
    },
    "like": {
        "Name": like_name,
        "sales": saleByLike,
        "profit": profitByLike
    }
}
data
```

```python
with open("./data/data.json", "w", encoding="utf-8")  as f:
    f.write(str(data))
```

```python
df8 = pd.pivot_table(df, values=["销售额"], index=["年份", "省市"], aggfunc="sum")
result = {}
for (year, city), value in df8["销售额"].items():
    if year not in result:
        result[year] = []
    result[year].append({"name": city, "value": value})
result
```
